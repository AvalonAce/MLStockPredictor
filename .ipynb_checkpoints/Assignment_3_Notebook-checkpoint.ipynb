```python
#import necessary libs
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
#use pandas to read the dataset
df = pd.read_csv('indian_liver_patient.csv')
```

```python
#clean the data
#remove missing vals
df = df.dropna()
```

```python
#encode categorical variables and scale features
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
#scale numerical columns
numerical_cols = ['Age', 'Total_Bilirubin', 'Direct_Bilirubin', 'Alkaline_Phosphotase',
                  'Alamine_Aminotransferase', 'Aspartate_Aminotransferase',
                  'Total_Protiens', 'Albumin', 'Albumin_and_Globulin_Ratio']
```

```python
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
X = df.drop('Dataset', axis=1)
y = df['Dataset']
```

```python
#balance the classes using SMOTE
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2)
```

```python
def eval_model(y_true, y_pred, model_name):
    #confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap='Blues')
    plt.title(f'{model_name} - Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    #calculate metrics
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='macro')
    recall = recall_score(y_true, y_pred, average='macro')
    f1 = f1_score(y_true, y_pred, average='macro')

    #print metrics
    print(f"{model_name} Performance:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print("-" * 30)
    
    return f1
```

## Random Forest Classifier

```python
#import lib
from sklearn.ensemble import RandomForestClassifier

#init the model with hyperparameters
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10)

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

#call eval_model func to get results
f1_rf = eval_model(y_test, y_pred_rf, 'Random Forest Classifier')
```

## Bagging Classifier

```python
#import libs
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

#init the base estimator
base_estimator = DecisionTreeClassifier()

#init bagging classifier with hyperparameters
bagging_model = BaggingClassifier(base_estimator=base_estimator, n_estimators=50, max_samples=0.8, bootstrap=True)

bagging_model.fit(X_train, y_train)

y_pred_bagging = bagging_model.predict(X_test)

#call eval_model func to get results
f1_bagging = eval_model(y_test, y_pred_bagging, 'Bagging Classifier')
```

## AdaBoost Classifier

```python
#import lib
from sklearn.ensemble import AdaBoostClassifier

#init the weak classifier
weak_classifier = DecisionTreeClassifier(max_depth=1)

#init the AdaBoost classifier w/ hyperparameters
adaboost_model = AdaBoostClassifier(base_estimator=weak_classifier, n_estimators=100, learning_rate=1)

adaboost_model.fit(X_train, y_train)

y_pred_adaboost = adaboost_model.predict(X_test)

#call eval_model func to get results
f1_adaboost = evaluate_model(y_test, y_pred_adaboost, 'AdaBoost Classifier')
```

## Stacking Classifier

```python
#import libs
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

#init base models
estimators = [
    ('dt', DecisionTreeClassifier()),
    ('knn', KNeighborsClassifier())
]

#init meta-classifier
meta_classifier = LogisticRegression()

#init Stacking Classifier
stacking_model = StackingClassifier(estimators=estimators, final_estimator=meta_classifier)

stacking_model.fit(X_train, y_train)

y_pred_stacking = stacking_model.predict(X_test)

#call eval_model func to get results
f1_stacking = evaluate_model(y_test, y_pred_stacking, 'Stacking Classifier')
```

## Model Evaluation

```python
#F1 Scores
results = pd.DataFrame({
    'Model': ['Random Forest', 'Bagging Classifier', 'AdaBoost Classifier', 'Stacking Classifier'],
    'F1 Score': [f1_rf, f1_bagging, f1_adaboost, f1_stacking]
})

print("Summary of Model Performance:")
print(results)

plt.figure(figsize=(8, 6))
sns.barplot(x='Model', y='F1 Score', data=results, palette='viridis')
plt.title('Model Comparison based on F1 Score')
plt.ylabel('F1 Score')
plt.xlabel('Model')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()
```
