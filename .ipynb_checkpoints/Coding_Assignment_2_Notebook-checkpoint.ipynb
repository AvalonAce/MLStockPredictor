```python
#Ryan Schreck

#import necessary libs
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

#use pandas to read the dataset
df = pd.read_csv('indian_liver_patient.csv')
```

```python
#explore the data
#display the first and last few rows
print(df.head())
print(df.tail())

print(df.info())

print(df.describe())

#check for null vals
print(df.isnull().sum())

#check for duplicates
print(df.duplicated().sum())

#check for anomalies in Total Bilirubin and Alkaline Phosphotase
sns.boxplot(data=df[['Total_Bilirubin', 'Alkaline_Phosphotase']])
plt.title('Boxplot of Total Bilirubin and Alkaline Phosphotase')
plt.show()

#check correlations
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#clean the data
#remove missing vals
df = df.dropna()
```

```python
#encode categorical variables and scale features
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

#scale numerical columns
numerical_cols = ['Age', 'Total_Bilirubin', 'Direct_Bilirubin', 'Alkaline_Phosphotase',
                  'Alamine_Aminotransferase', 'Aspartate_Aminotransferase',
                  'Total_Protiens', 'Albumin', 'Albumin_and_Globulin_Ratio']

scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
#visualize the class imbalance
print(df['Dataset'].value_counts())
```

```python
#plot the imbalance
sns.countplot(x='Dataset', data=df)
plt.title('Class Imbalance')
plt.show()
```

```python
#balance the classes using SMOTE
X = df.drop('Dataset', axis=1)
y = df['Dataset']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
smote = SMOTE()
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
#logisitic regression
lr = LogisticRegression()
lr.fit(X_train_resampled, y_train_resampled)
```

```python
#Decision Tree
dt = DecisionTreeClassifier(criterion='entropy', max_depth=4)
dt.fit(X_train_resampled, y_train_resampled)
```

```python
#KNN
knn = KNeighborsClassifier(weights='distance', algorithm='kd_tree')
knn.fit(X_train_resampled, y_train_resampled)
```

```python
#KNN w/ GridSearchCV
param_grid = {'n_neighbors': [3, 5, 7, 9], 'weights': ['uniform', 'distance'], 'algorithm': ['auto', 'ball_tree', 'kd_tree']}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid.fit(X_train_resampled, y_train_resampled)
```

```python
#predictions
y_pred_lr = lr.predict(X_test)
y_pred_dt = dt.predict(X_test)
y_pred_knn = knn.predict(X_test)
y_pred_knn_grid = grid.predict(X_test)
```

```python
#eval each model
#func to eval the model
def evaluate_model(y_test, y_pred, model_name):
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d")
    plt.title(f'{model_name} Confusion Matrix')
    plt.show()

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro')
    recall = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')

    print(f"{model_name} - Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")
    return f1

#eval Logistic Regression
f1_lr = evaluate_model(y_test, y_pred_lr, 'Logistic Regression')

#eval Decision Tree
f1_dt = evaluate_model(y_test, y_pred_dt, 'Decision Tree')

#eval KNN
f1_knn = evaluate_model(y_test, y_pred_knn, 'KNN')

#eval KNN with GridSearchCV
f1_knn_grid = evaluate_model(y_test, y_pred_knn_grid, 'KNN (GridSearch)')

#compare models based on F1 Score
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Decision Tree', 'KNN', 'KNN GridSearch'],
    'F1 Score': [f1_lr, f1_dt, f1_knn, f1_knn_grid]
})

print(results)
```

```python
print('These scores seem to imply that the models are balancing precision and recall moderately and could be improved upon. All of the scores tend to be similar in each model.')
```

```python
#Part 2
#implement logistic regression from scratch

#sigmoid func
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

#cost func
def compute_cost(X, y, weights):
    m = len(y)
    h = sigmoid(X @ weights)
    cost = (-1/m) * (y.T @ np.log(h) + (1 - y).T @ np.log(1 - h))
    return cost

#gradient descent
def gradient_descent(X, y, weights, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        h = sigmoid(X @ weights)
        gradient = (1/m) * (X.T @ (h - y))
        weights -= learning_rate * gradient
        cost = compute_cost(X, y, weights)
        cost_history.append(cost)

    return weights, cost_history

```

```python
#add intercept term
X_train.insert(0, 'Intercept', 1)
X_test.insert(0, 'Intercept', 1)

weights = np.zeros(X_train.shape[1])

learning_rate = 0.001
iterations = 2000
```

```python
#train the model using gradient descent
weights, cost_history = gradient_descent(X_train.values, y_train.values, weights, learning_rate, iterations)
```

```python
#make predictions
y_pred_prob = sigmoid(X_test.values @ weights)
y_pred = [2 if prob >= 0.7 else 1 for prob in y_pred_prob]
```

```python
print("Class distribution in y_test:", pd.Series(y_test).value_counts())
print("Class distribution in y_pred:", pd.Series(y_pred).value_counts())
```

```python
#eval the model
f1_log_reg_scratch = evaluate_model(y_test, y_pred, 'Logistic Regression (Scratch)')
```

```python
print('These results indicate that the model holds a bias towards predicting class 2. After experimnenting with many combinations of learning rates and iterations, this remains to be the case. The model has a low accuracy score and a moderate precision score. All of this is reflected in the low F1 score. Experimenting with differing combinations of learning rate and iterations yielded an F1 score of 0.23 in the worst case, and 0.37 in the best case.')
```

```python
#plotting the cost func against the number of iterations
plt.plot(range(iterations), cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost Function Convergence')
plt.show()
```
