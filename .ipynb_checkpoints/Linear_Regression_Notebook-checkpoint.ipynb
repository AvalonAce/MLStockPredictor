```python
#Ryan Schreck
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
%matplotlib inline
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('housing_data_new.csv')
```

```python
print("First 5 rows:")
display(df.head())

print("\nLast 5 rows:")
display(df.tail())
```

```python
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
numerical_features = [col for col in df.columns if col not in ['Basement', 'Hot Water Heating', 'Air Conditioning', 'Mainroad', 'Garage', 'Price']]
df[numerical_features].hist(bins=30, figsize=(15, 10))
plt.suptitle('Histogram of Numerical Features')
plt.show()
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    initial_shape = df.shape
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    final_shape = df.shape
    print(f"Removed outliers from '{column}': {initial_shape[0] - final_shape[0]} rows")
    return df

df = remove_outliers_iqr(df, 'Size')
df = remove_outliers_iqr(df, 'Age')

print(f"\nDataset shape after removing outliers: {df.shape}")
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
features_to_fill = df.columns.drop('Price')
df[features_to_fill] = df[features_to_fill].fillna(df[features_to_fill].median())

print("\nMissing values after imputation:")
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
print(f"\nDataset shape after dropping duplicates: {df.shape}")
```

```python
binary_features = ['Basement', 'Hot Water Heating', 'Air Conditioning', 'Mainroad', 'Garage']

numerical_features = [col for col in df.columns if col not in binary_features + ['Price']]

scaler = StandardScaler()
X = df.drop('Price', axis=1)
y = df['Price']

X_scaled = X.copy()
X_scaled[numerical_features] = scaler.fit_transform(X[numerical_features])

print("\nScaled Features:")
display(X_scaled.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)

print(f"\nTraining set size: {X_train.shape}")
print(f"Testing set size: {X_test.shape}")
```

```python
lr_model = LinearRegression()

lr_model.fit(X_train, y_train)
print("\nModel training complete.")
```

```python
y_pred = lr_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
print(f"\nMean Absolute Error (MAE): {mae:.2f}")
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
```

```python
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2): {r2:.2f}")
```

```python
print("\nExplaining the model's results:")
print(f"The R-squared value of {r2:.2f} indicates that approximately {r2*100:.2f}% of the variability in house prices is explained by the model's features. The MAE and RMSE give us information regarding the average prediction errors, with lower values indicating better performance.")
```

```python
cv_scores = cross_val_score(lr_model, X_scaled, y, cv=5, scoring='r2')

print(f"\nCross-Validation R2 Scores: {cv_scores}")
print(f"Average Cross-Validation R2 Score: {cv_scores.mean():.2f}")
```

```python
def hypothesis(X, theta):
    return np.dot(X, theta)

def compute_cost(X, y, theta):
    m = len(y)
    predictions = hypothesis(X, theta)
    cost = (1/(2*m)) * np.sum((predictions - y) ** 2)
    return cost

def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        predictions = hypothesis(X, theta)
        errors = predictions - y
        gradient = (1/m) * np.dot(X.T, errors)
        theta = theta - learning_rate * gradient
        cost = compute_cost(X, y, theta)
        cost_history.append(cost)
        
        if i % 100 == 0:
            print(f"Iteration {i}: Cost {cost:.4f}")

    return theta, cost_history
```

```python
X_train_np = X_train.values
X_test_np = X_test.values

X_train_b = np.hstack((np.ones((X_train_np.shape[0], 1)), X_train_np))
X_test_b = np.hstack((np.ones((X_test_np.shape[0], 1)), X_test_np))

#converting targets to numpy arrays
y_train_np = y_train.values
y_test_np = y_test.values

#init theta to zeros
theta_initial = np.zeros(X_train_b.shape[1])

#set learning rate and iterations
learning_rate = 0.01
iterations = 1000

#perform gradient descent
theta_final, cost_history = gradient_descent(X_train_b, y_train_np, theta_initial, learning_rate, iterations)

print("\nFinal theta parameters:", theta_final)
```

```python
def predict(X, theta):
    return hypothesis(X, theta)
```

```python
y_pred_scratch = predict(X_test_b, theta_final)

#Eval metrics
#Mean Absolute Error
mae_scratch = mean_absolute_error(y_test_np, y_pred_scratch)
print(f"\nMAE (Scratch): {mae_scratch:.2f}")

#Mean Squared Error
mse_scratch = mean_squared_error(y_test_np, y_pred_scratch)
print(f"MSE (Scratch): {mse_scratch:.2f}")

#Root Mean Squared Error
rmse_scratch = np.sqrt(mse_scratch)
print(f"RMSE (Scratch): {rmse_scratch:.2f}")

#R-squared
ss_total = np.sum((y_test_np - np.mean(y_test_np)) ** 2)
ss_residual = np.sum((y_test_np - y_pred_scratch) ** 2)
r2_scratch = 1 - (ss_residual / ss_total)
print(f"R-squared (Scratch): {r2_scratch:.2f}")
```

```python
#Plotting the cost func against the # of iterations
plt.figure(figsize=(6,3))
plt.plot(range(iterations), cost_history, color='blue')
plt.title('Cost Function vs. Iterations')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

```python
print(f"The plot appears to show a smooth curve. Over time, the rate of decrease slows down and starts to flatten. This shows that the algo moves towards the minimum cost and stabalizes.")
```
